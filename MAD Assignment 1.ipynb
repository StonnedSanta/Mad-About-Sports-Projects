### Strike Rate of 158.04 by MS Dhoni in Death Overs is quite impressive as compared mainly to middle overs.
### Lets find out the Strike Rate of other players 

```python
df.head()
```

```python
df1 = pd.DataFrame([df.match_id, df.season, df.striker, df.runs_off_bat, df.ball, df.phase]).transpose()
```

```python
df1.sample(6)
```

```python
df1 = df1.query("phase == 'DO'")
```

```python
df1
```

```python
adf = pd.DataFrame(df1.groupby('striker')['runs_off_bat'].sum())
```

```python
bdf = pd.DataFrame(df1.groupby('striker')['ball'].count())
```

```python
cdf = adf.merge(bdf, on = 'striker', how = 'left')
```

```python
cdf = cdf[(cdf.runs_off_bat >= 150) & (cdf.ball >= 150)].reset_index()
```

```python
cdf['strike_rate'] = 100 * cdf.runs_off_bat / cdf.ball
```

```python
cdf.head(5)
```

```python
cdf.sort_values('strike_rate', ascending = False).reset_index()
```
